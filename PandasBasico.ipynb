# **Significado de Pandas**

Pandas é uma ferramenta muito poderosa para análise de dados usando dados reais.

Pandas é rápido, pois usa Numpy, que é escrito em C.

*DataFrame*

Os dados no pandas, primeiro precisamos preparar os dados para tipos de dicionários e compreensões de listas. Um tipo de dicionário é um tipo de dados onde a chave e o valor são emparelhados. No exemplo a seguir, definimos as Chaves 'Ano' e 'Índice' e seus correspondentes tipos de lists Valores.

A funçãol usada gerando uma sequencia de números inteiros num intervalo especifico. Passar um único argumento produz uma lista de números inteiros que são incrementados em 1 de 0 (o tamanho da lista é gerado para o número especificado no argumento). Se você passar dois argumentos, você pode especificar o início e o fim da lista mais um.

Exemplo: [i para i no intervalo(1,3)] -> para gerar: [1,2]

```python
import pandas as pd
year_index = {'Year': [y for y in range(2010, 2031)],
              'Index': [i for i in range(21)],
              }

print(year_index)
```

O tipo mais comum usado em pandas é o tipo DataFrame, seguido pelo tipo Series e, dependendo do trabalho, o tipo Painel, que não é usado tanto.

Veremos apenas o DataFrame:

```python
df = pd.DataFrame(year_index)
print(df.head())
```

```python
df.head(10)
```

```python
df.sample(5)
```

```python
import random
years_random = {'Year': [y for y in range(2010, 2031)],
                'Random': [random.random() for y in range(21)]
                }

df1 = pd.DataFrame(years_random)

df1.head()
```

**Visualização apenas com pandas**

Pandas também faz uso de matplotlib internamente. portanto, o tipo DataFrame tem um método chamado "plot" que pode facilmente visualizar os dados armazenados nele. Por exemplo, se você quiser exibir um gráfico de dispersão de df1 com Ano no eixo x e random no eixo y.

```python
df1.plot(x = df1.columns[0], y = df1.columns[1], kind = 'scatter', xticks = [year for year in range(2010, 2031, 5)])
```

**Tabela com chave específica**

As mesas de junção: VLOOOKUP e XLOOOKUP do EXCEL também são úteis , mas quando temos grandes quantidades de dados há real possibilidade de travamentos e até de imposibilidade de uso.

Já com pandas, você pode unir tabelas facilmente com colunas especificas como as chaves...

```python
years_random2 = {'Year': [y for y in range(2011, 2032)],
                'Random': [random.random() for y in range(21)]
                }

df2 = pd.DataFrame(years_random2)
```

Vamos pegar o que há de comum nos dois pelas chaves, usando a união interna(inner).

```python
pd.merge(df1, df2, on = 'Year', how = 'inner')
```

Agora pelo ano, usando a externa(outer).

```python
pd.merge(df1, df2, on = 'Year', how = 'outer')
```

Agora juntas exteriores esquerda..., neste caso df1 tera seus dados preservados, sem NaN.

```python
pd.merge(df1, df2, on = 'Year', how = 'left')
```

O mesmo para direita

```python
pd.merge(df1, df2, on = 'Year', how = 'right')
```

# **Numpy**

Numpy é uma biblioteca muito importante na qual se baseia outras bibliotecas de análise de dados(por exemplo, Pandas/Matplotlib/SciPy), e é uma das principais razões pelas quais Python se tornou uma linguagem tão importante. Numpy permite operações de dados super-rápidas em arrays multidimensionais.

Numpy usa um formato de retenção de dados chamada ndarray, que ao contrário das listas de python acima mencionadas, só pode armazenar dados do mesmo tipo em ndarray, mas em vez disso aloca os dados continuamente na memória (RAM). Ao armazenar dados em uma área continua da memória, os dados podem ser lidos eficientemente em registros de CPU.

Outra vantagem é que o Numpy são as bibliotecas lineares BLAS e LAPACK, que realizam operações de vectorização na CPU em tempo de compilação, para ter beneficio das operações de alta velocidade.

```python
import numpy as np

np.__config__.show()
```

Vamos usar os tipos de lista do Python e como o ndarray do Numpy é armazenado na memória, usando o poder do CPython.

```python
import random

list1 = [random.random() for i in range(10)]
np_arr = np.array(list1)
```

Agora exibimos na ordem do indice.

```python
for i, elm in enumerate(np_arr):
  print(f'index do enumerate: {i}, id do np_arr: {id(np_arr[i])}, Elemento: {elm}')
```

Os valores do id se repetem, pois o ndarray armazena em uma atribuição continua de espaço.

```python
for i, elm in enumerate(list1):
  print(f'index do enumerate: {i}, id do list1: {list1[i]}, Elemento: {elm}')
```

Já na lista o Python tipa os mesmos de forma dinamica, o que obviamente faz com que os mesmos sejam mais lentos.

# **Função de Geração Numpy Array**

A geração de matrizes multidimensionais é a caracteristica determinante do Numpy.



```python
import numpy as np

z = np.zeros((3,3))
print(z)
```

Numpy.zeros gera um array com elementos 0, cada elemento de 8 bytes e do tipo float64.

```python
def print_label_val(label, val):
  print(f'{label} : {val}')

def print_ndr_basic_info(z):
  print_label_val('type: ', type(z))
  print_label_val('shape: ', z.shape)
  print_label_val('itemsize: ', z.itemsize)
  print_label_val('size: ', z.size)
  print_label_val('dtype: ', z.dtype)
  print_label_val('ndim: ', z.ndim)

print_ndr_basic_info(z)
```

Usando Numpy.arange vamos criar um array unidimensional com um elemento inicial e um final de +1, e o Numpy.reshape que permite modificar a forma, portanto o seguinte código define um arrray tridimensional de 2 x 3 x 4 pode ser feito.

```python
z = np.arange(0,24).reshape(2,3,4)
z
```
